## ) Problem 2: Bonus Calculation Based on Department and Performance

```python
%sql
SELECT 
    emp_name,
    department,
    salary,
    performance_rating,
    CASE 
        WHEN department = 'Finance' THEN 
            CASE 
                WHEN performance_rating = 'A' THEN salary * 0.20
                WHEN performance_rating = 'B' THEN salary * 0.15
                WHEN performance_rating = 'C' THEN salary * 0.05
            END

        WHEN department = 'Engineering' THEN 
            CASE 
                WHEN performance_rating = 'A' THEN salary * 0.18
                WHEN performance_rating = 'B' THEN salary * 0.12
                WHEN performance_rating = 'C' THEN salary * 0.03
            END

        ELSE salary * 0.10
    END AS bonus
FROM Employee;
```

## Problem 3: Categorizing Employees by Salary Range and Performance

```python
%sql
select emp_name,salary,
case 
    when salary >=80000 and performance_rating ='A'then 'High Performer'
    when salary >=50000 and salary <80000 and performance_rating ='B'then 'Average Performer'
    when salary <50000 and performance_rating ='C' then 'Low Performer'
end AS salary
FROM Employee
```

```python
%sql
SELECT 
    emp_name,
    department,
    experience,
    CASE 
        -- HR Department
        WHEN department = 'HR' AND experience < 5 THEN 'High Risk'
        WHEN department = 'HR' AND experience >= 5 THEN 'Low Risk'
        
        -- Engineering & Finance
        WHEN department IN ('Engineering', 'Finance') AND experience > 8 THEN 'Low Risk'
        WHEN department IN ('Engineering', 'Finance') AND experience <= 8 THEN 'Medium Risk'
        
        -- Other Departments
        ELSE 'Medium Risk'
    END AS risk_level
FROM Employee;
```
