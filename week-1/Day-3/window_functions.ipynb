## 1.Assign a unique row number to all employees based on salary (highest first). 

```python
%sql
SELECT 
    *,
    ROW_NUMBER() OVER (ORDER BY salary DESC) AS row_num
FROM employees;
```

## Assign row numbers to employees within each department based on salary descending.

```python
%sql
select *,
row_number() over(partition by department order by salary desc) as row_num
from employees
```

```python
%sql
SELECT 
    emp_id,
    emp_name,
    join_date,
    ROW_NUMBER() OVER (ORDER BY join_date DESC) AS row_num
FROM employees;
```

## 4.Assign row numbers within each department based on earliest joining date. 

```python
%sql
SELECT 
    emp_id,
    emp_name,
    department,
    join_date,
    ROW_NUMBER() OVER (
        PARTITION BY department 
        ORDER BY join_date ASC
    ) AS row_num
FROM employees;
```

## Rank all employees based on salary (highest first).

```python
%sql
SELECT 
    emp_id,
    emp_name,
    salary,
    RANK() OVER (ORDER BY salary DESC) AS rank_num
FROM employees;
```

## Rank employees within each department based on salary. 

```python
%sql
SELECT 
    emp_id,
    emp_name,
    department,
    salary,
    RANK() OVER (
        PARTITION BY department 
        ORDER BY salary DESC
    ) AS rank_num
FROM employees;
```

## Rank employees based on joining date (latest gets rank 1). 

```python
%sql
SELECT 
    emp_id,
    emp_name,
    join_date,
    RANK() OVER (ORDER BY join_date DESC) AS rank_num
FROM employees;
```

## Assign dense rank to employees based on salary (highest first). 

```python
%sql
SELECT 
    emp_id,
    emp_name,
    salary,
    DENSE_RANK() OVER (ORDER BY salary DESC) AS dense_rank_num
FROM employees;
```
