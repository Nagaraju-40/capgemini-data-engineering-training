```python
%sql
CREATE TABLE Employee (
emp_id INT PRIMARY KEY,
emp_name VARCHAR(50),
department VARCHAR(50),
salary DECIMAL(10, 2),
joining_date DATE
);
INSERT INTO Employee (emp_id, emp_name, department, salary, joining_date) VALUES
(1, 'Karthik', 'HR', 60000.00, '2021-01-15'),
(2, 'Pratik', 'Finance', 70000.00, '2021-03-10'),
(3, 'Veer', 'HR', 55000.00, '2021-06-20'),
(4, 'Priya', 'Finance', 80000.00, '2022-01-05'),
(5, 'Ajay', 'Engineering', 75000.00, '2020-11-01'),
(6, 'Vijay', 'Engineering', 78000.00, '2019-05-22'),
(7, 'Veena', 'HR', 62000.00, '2023-03-12'),
(8, 'Meena', 'Marketing', 65000.00, '2022-08-18');
```

```python
%sql
--  Find the total salary for each department in the Employee table.
SELECT department, SUM(salary) AS total_salary
FROM Employee
GROUP BY department;
```

```python
%sql
SELECT department, COUNT(*) AS number_of_employees
FROM Employee
GROUP BY department;
```

```python
%sql
SELECT department, AVG(salary) AS avg_salary
FROM Employee
GROUP BY department;
```

```python
%sql
SELECT department, MAX(salary) AS MAX_salary
FROM Employee
GROUP BY department;
```

```python
%sql
SELECT department, MIN(salary) AS MIN_salary
FROM Employee
GROUP BY department;
```

```python
%sql
-- Find the total salary for departments where the total salary exceeds 100,000.
SELECT department, SUM(salary) AS total_salary
FROM Employee
GROUP BY department
HAVING SUM(salary )>100000
```

##. Count the number of employees in departments with more than 2 employees

```python
%sql
SELECT department, count(*) num_employees
FROM Employee
GROUP By department
HAVING count(*)>2
```

## Calculate the average salary for employees who joined after 2021-01-01, grouped by department.


```python
%sql
SELECT department, AVG(salary) AS avg_salary  FROM employee WHERE joining_date>'2021-01-01'
GROUP BY department
```

##. Find the departments where the maximum salary is greater than 75,000.


```python
%sql
SELECT department,MAX(salary) AS max_salary
FROM Employee
WHERE salary>75000
GROUP BY department
```

## List the departments where the total salary is less than 150,000.

```python
%sql
SELECT department, SUM(salary) AS total_salary 
FROM Employee
WHERE salary<150000
GROUP BY department;
```

##11.Find the total number of employees grouped by department, but only include departments with more than 1 employee.

```python
%sql
SELECT department, COUNT(*) AS total_employees
FROM Employee
GROUP BY department
HAVING COUNT(*) > 1;
```

##12. Calculate the total salary of each department and show only those where the total exceeds 125,000.

```python
%sql
SELECT department, SUM(salary) AS total_salary
FROM Employee
GROUP BY department
HAVING total_salary > 125000;
```

## 13. Count the number of employees in each department, but include only departments with more than 2 employees.

```python
%sql
select count(emp_id),department from employee
group by department
having count(emp_id) > 2
```

## 14. Find the average salary in each department where the average salary is above 60,000.

```python
%sql
select department,avg(salary) as Avg_Salary from employee
group by department
having avg(salary) > 60000
```
